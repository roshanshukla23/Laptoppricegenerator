```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
a=pd.read_csv("/content/drive/MyDrive/MATPLOTLIB/pandas/datasets/knn.csv")
a

```

```python
a['Height(y2)cm']=a['Height(y2)cm'].apply(lambda a:a.replace("cm","")).astype('int')
```

```python
a.info()
```

```python
a
```

```python
a['Weight(x2)kg']=a['Weight(x2)kg'].apply(lambda a:a.replace("kg","")).astype('int')
```

```python
a
```

```python
a.info()
```

```python
 a.duplicated().sum
```

```python
X=a.iloc[:,[0,1]].values
y=a.iloc[:,2].values
k=5
from sklearn.neighbors import KNeighborsClassifier
```

```python
model=KNeighborsClassifier(n_neighbors=k,metric='euclidean')
```

```python
model.fit(X,y)
```

```python
model.predict([[65.2,175]])
```

```python
df=a
df1=df[df['Class']=='Normal']
df2=df[df['Class']=='Underweight']
df3=df[df['Class']=='Overweight']
```

```python
df1
```

```python
df2
```

```python
df3
```

```python

plt.scatter(df1['Weight(x2)kg'],df1['Height(y2)cm'],label="Normal",marker="*")
plt.scatter(df3['Weight(x2)kg'],df3['Height(y2)cm'],label="Overweight",marker="*")
plt.scatter(df2['Weight(x2)kg'],df2['Height(y2)cm'],label="Underweight",marker="*")
plt.legend()
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
y_pred=model.predict(X)
```

```python
y_pred
```

```python
y
```

```python
accuracy_score(y,y_pred)
```

```python
pd.DataFrame(confusion_matrix(y,y_pred),index=['normal_A','overweight_A','underweight_A'],
             columns=['normal_B','overweight_B','underweight_B'])
```

```python
confusion_matrix(y,y_pred)
```

```python
print(classification_report(y,y_pred))
```

```python
%%writefile app.py
import streamlit as st
import joblib
st.title("Prediction of Class based on Weight and Height ")
st.text("Please select the values for weight and heightand click on PREDICT")

model=joblib.load('knn_model')

weight=st.slider("Weight(in kg)",min_value=40,max_value=90,value=55,step=1)
height=st.slider("Height(in cm)",min_value=150,max_value=200,value=160,step=1)

if st.button("PREDICT"):
  op=model.predict([[weight,height]])
  st.subheader("the person having a weight of "+str(weight)+"kg and height of "+str(height)+"cm is considered as "+op[0]+".")
```
